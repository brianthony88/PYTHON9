```python
# We only want the Driver Standings
# for x in range(5):    
#     data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]['Driver']['familyName']
#     print(data)

my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} [{racer['Driver']['familyName']}]"
        
        racer_dict[racer_name] = {
            'first_name': racer['Driver']['givenName'],
            'last_name': racer['Driver']['familyName'],
            'position': racer['position'],
            'wins': racer['wins'],
            'DOB': racer['Driver']['dateOfBirth'],
            'nationality': racer['Driver']['nationality'],
            'constructor': racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(my_racer_data)
```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f'http://ergast.com/api/f1/{year}/{r}/driverStandings.json'
    response = requests.get(url)
    if not response.json()['MRData']['StandingsTable']['StandingsLists']:
        return "We had an error loading your data. It is likely the year or round is not in the database."
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    return get_racer_info(data)

get_driver_info_by_year_and_round(2014,4)
```

```python
# THIS IS FOR SEARCHING THE 20 ON THE LIST THE SITE HAS. THE FIRST 20
# import requests

# # Function to fetch data for a single Pokémon
# def get_pokemon_data(pokemon_id):
#     url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_id}"
#     response = requests.get(url)
#     if response.status_code == 200:
#         pokemon_data = response.json()
#         pokemon_name = pokemon_data['name'].capitalize()
#         ability_name = pokemon_data['abilities'][0]['ability']['name']
#         base_experience = pokemon_data['base_experience']
#         sprite_url = pokemon_data['sprites']['front_shiny']
#         attack_stat = pokemon_data['stats'][1]['base_stat']
#         hp_stat = pokemon_data['stats'][0]['base_stat']
#         defense_stat = pokemon_data['stats'][2]['base_stat']
#         return {
#             'name': pokemon_name,
#             'ability': ability_name,
#             'base_experience': base_experience,
#             'sprite_url': sprite_url,
#             'attack_stat': attack_stat,
#             'hp_stat': hp_stat,
#             'defense_stat': defense_stat
#         }
#     else:
#         print(f"Failed to fetch data for Pokemon ID {pokemon_id}. Status code: {response.status_code}")
#         return None

# # Function to fetch data for multiple Pokémon
# def get_multiple_pokemon_data(num_pokemon):
#     pokemon_data_list = []
#     for i in range(1, num_pokemon + 1):
#         pokemon_data = get_pokemon_data(i)
#         if pokemon_data:
#             pokemon_data_list.append(pokemon_data)
#     return pokemon_data_list

# # Fetch data for 20 Pokémon
# pokemon_data = get_multiple_pokemon_data(20)

# # Print the data for each Pokémon
# for data in pokemon_data:
#     print("Name:", data['name'])
#     print("Ability:", data['ability'])
#     print("Base Experience:", data['base_experience'])
#     print("Sprite URL:", data['sprite_url'])
#     print("Attack Base Stat:", data['attack_stat'])
#     print("HP Base Stat:", data['hp_stat'])
#     print("Defense Base Stat:", data['defense_stat'])
#     print("-" * 50)


# THIS IS FOR SPECIFIC NAMES YOU WANT TO USE/SEARCH. 
import requests

# Function to fetch data for a single Pokémon
def get_pokemon_data(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
    response = requests.get(url)
    if response.status_code == 200:
        pokemon_data = response.json()
        pokemon_name = pokemon_data['name'].capitalize()
        ability_name = pokemon_data['abilities'][0]['ability']['name']
        base_experience = pokemon_data['base_experience']
        sprite_url = pokemon_data['sprites']['front_shiny']
        attack_stat = pokemon_data['stats'][1]['base_stat']
        hp_stat = pokemon_data['stats'][0]['base_stat']
        defense_stat = pokemon_data['stats'][2]['base_stat']
        return {
            'name': pokemon_name,
            'ability': ability_name,
            'base_experience': base_experience,
            'sprite_url': sprite_url,
            'attack_stat': attack_stat,
            'hp_stat': hp_stat,
            'defense_stat': defense_stat
        }
    else:
        print(f"Failed to fetch data for Pokemon {pokemon_name}. Status code: {response.status_code}")
        return None

# List of specific Pokémon names
pokemon_names = [
    "psyduck", "gyarados", "machamp", "ninetales", "gloom",
    "dugtrio", "primeape", "poliwhirl", "alakazam", "tentacruel",
    "ponyta", "magnemite", "dewgong", "gengar", "onix",
    "electrode", "exeggcute", "marowak", "weezing", "starmie"
]

# Fetch data for specific Pokémon
pokemon_data_list = []
for name in pokemon_names:
    pokemon_data = get_pokemon_data(name)
    if pokemon_data:
        pokemon_data_list.append(pokemon_data)

# Print the data for each Pokémon
for data in pokemon_data_list:
    print("Name:", data['name'])
    print("Ability:", data['ability'])
    print("Base Experience:", data['base_experience'])
    print("Sprite URL:", data['sprite_url'])
    print("Attack Base Stat:", data['attack_stat'])
    print("HP Base Stat:", data['hp_stat'])
    print("Defense Base Stat:", data['defense_stat'])
    print("-" * 50)

```
